```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
data = pd.read_csv("Data/Uber-Jan-Feb-FOIL.csv")
data['date'] = pd.to_datetime(data['date'])
data = data.sort_values('date')
```

```python
data['day_of_week'] = data['date'].dt.day_name()
data['is_weekend'] = data['day_of_week'].isin(['Saturday','Sunday']).astype(int)
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
data['trips_rolling_mean_3'] = data['trips'].rolling(window=3).mean()
data['trips_rolling_mean_7'] = data['trips'].rolling(window=7).mean()
data['lag_1'] = data['trips'].shift(1)
data['lag_2'] = data['trips'].shift(2)
data['lag_3'] = data['trips'].shift(3)
data = data.dropna().reset_index(drop=True)
```

```python
X = data[['active_vehicles', 'is_weekend', 'month', 'day',
          'trips_rolling_mean_3', 'trips_rolling_mean_7',
          'lag_1', 'lag_2', 'lag_3']]
y = data['trips']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)


models = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "XGBoost": XGBRegressor(random_state=42, n_estimators=200),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    mape = mean_absolute_percentage_error(y_test, preds) * 100
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    r2 = r2_score(y_test, preds)
    results[name] = [mape, rmse, r2]
```

```python
results_df = pd.DataFrame(results, index=["MAPE (%)", "RMSE", "R² Score"]).T
display(results_df)
```

```python
best_model = min(results.items(), key=lambda x: x[1][0])[0]  # model with lowest MAPE
print(f"\nBest Model: {best_model}")

best = models[best_model]
best_preds = best.predict(X_test)

plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual Trips", color="blue")
plt.plot(best_preds, label="Predicted Trips", color="orange")
plt.title(f"Actual vs Predicted Trips ({best_model})")
plt.xlabel("Test Data Points")
plt.ylabel("Trips")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fix: Convert best_model name to actual model object
best_model_name = min(results.items(), key=lambda x: x[1][0])[0]
best_model = models[best_model_name]

print("✅ Best model selected:", best_model_name)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Feature importance from Gradient Boosting
importances = best_model.feature_importances_
features = X.columns

fi = pd.DataFrame({'Feature': features, 'Importance': importances})
fi = fi.sort_values('Importance', ascending=True)

# Plot
plt.figure(figsize=(8, 5))
plt.barh(fi['Feature'], fi['Importance'], color='skyblue')
plt.title("Feature Importance (Gradient Boosting)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.grid(True)
plt.tight_layout()
plt.show()

fi
```

```python
import shap
import numpy as np

# Use TreeExplainer for tree-based models (fast + accurate)
explainer = shap.TreeExplainer(best_model, X_train)   # best_model must be the trained model object
shap_values = explainer.shap_values(X_test)           # returns array of SHAP values

# Quick check
print("SHAP values type:", type(shap_values))
try:
    print("shape:", np.array(shap_values).shape)
except Exception:
    pass
```

```python
import matplotlib.pyplot as plt

# 1) Bar summary (global importance)
plt.figure(figsize=(8,5))
shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
plt.title("SHAP Feature Importance (bar)")
plt.tight_layout()
plt.savefig("shap_summary_bar.png", dpi=150)
plt.show()
print("Saved: shap_summary_bar.png")

# 2) Beeswarm (detailed per-sample effects)
plt.figure(figsize=(10,6))
shap.summary_plot(shap_values, X_test, plot_type="dot", show=False, max_display=12)
plt.title("SHAP Summary (beeswarm)")
plt.tight_layout()
plt.savefig("shap_beeswarm.png", dpi=150)
plt.show()
print("Saved: shap_beeswarm.png")
```

```python
i = 5   # change to any test index you want to inspect
print("Explaining test index:", i)

# Interactive JS widget (shows inline in notebook)
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[i], X_test.iloc[i], matplotlib=True, show=True)

# Save interactive HTML in the same folder as your notebook
force_html = shap.force_plot(explainer.expected_value, shap_values[i], X_test.iloc[i]).html()
file_path = f"shap_force_index_{i}.html"
with open(file_path, "w", encoding="utf-8") as f:
    f.write(force_html)

print(f"✅ Saved interactive explanation as: {file_path}")
```

```python
Insights & Key Learnings

The Gradient Boosting Regressor achieved R² = 0.983 and MAPE ≈ 7.1%, indicating exceptional predictive accuracy.

Active vehicles, recent trip averages, and previous day’s trips were the most influential factors driving trip volume.

Weekend flag and day of the month had smaller but consistent impacts.

The SHAP analysis confirmed that:

Higher active vehicles → more trips.

Weekends slightly increase predicted trip demand.

Rolling 3-day and 7-day averages stabilize predictions, capturing short-term patterns.

The model can help Uber optimize driver allocation and demand forecasting, reducing under- or over-supply in specific zones.
```
