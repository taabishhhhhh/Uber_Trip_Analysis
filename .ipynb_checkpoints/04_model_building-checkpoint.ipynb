```python
""" 04 ‚Äî Model Training, Evaluation & Interpretability 
This notebook loads the processed train/test datasets, trains ML models, compares performance, selects the best model, analyzes feature importance, generates SHAP explainability plots, and saves the best model."""
```

```python
import joblib
import pandas as pd

try:
    X_train = joblib.load("models/X_train.joblib")
    X_test  = joblib.load("models/X_test.joblib")
    y_train = joblib.load("models/y_train.joblib")
    y_test  = joblib.load("models/y_test.joblib")
    print("Loaded train/test splits from models/.")
except Exception as e:
    print("ERROR loading splits:", e)
    raise SystemExit("Train/Test splits missing ‚Äî run 03_train_test_split.ipynb first!")
```

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score

models = {
    "Random Forest": RandomForestRegressor(random_state=42, n_estimators=200, n_jobs=-1),
    "XGBoost": XGBRegressor(random_state=42, n_estimators=200, verbosity=0),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42, n_estimators=200)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    results[name] = {
        "MAPE (%)": mean_absolute_percentage_error(y_test, preds) * 100,
        "RMSE": np.sqrt(mean_squared_error(y_test, preds)),
        "R2": r2_score(y_test, preds)
    }

print("Training complete.")
```

```python
import pandas as pd

results_df = pd.DataFrame(results).T
results_df_rounded = results_df.round({'MAPE (%)':3, 'RMSE':2, 'R2':3})

print("Raw metrics:")
display(results_df)

print("\nRounded metrics:")
display(results_df_rounded)
```

```python
best_model_name = results_df['MAPE (%)'].idxmin()
best_model = models[best_model_name]

print(f"üèÜ Best Model: {best_model_name}")
print(f"MAPE: {results_df.loc[best_model_name,'MAPE (%)']:.3f}%")
```

```python
import matplotlib.pyplot as plt

preds = best_model.predict(X_test)

plt.figure(figsize=(12,6))
plt.plot(y_test.values, label="Actual Trips", color="blue", marker='o')
plt.plot(preds, label=f"Predicted ({best_model_name})", color="orange", linestyle="--", marker='o')
plt.title(f"Actual vs Predicted Trips ‚Äî {best_model_name}")
plt.xlabel("Test index (time-ordered)")
plt.ylabel("Trips")
plt.legend()
plt.grid(True)
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

if hasattr(best_model, "feature_importances_"):
    importances = best_model.feature_importances_
    features = X_train.columns

    fi = pd.DataFrame({'Feature': features, 'Importance': importances})
    fi = fi.sort_values('Importance', ascending=True)

    plt.figure(figsize=(8,5))
    plt.barh(fi['Feature'], fi['Importance'], color='skyblue')
    plt.title("Feature Importance")
    plt.xlabel("Importance Score")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    display(fi)
else:
    print("This model does not provide feature_importances_.")
```

```python
!pip install shap --quiet
import shap
import numpy as np

explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test)

print("SHAP ready. Shape:", np.array(shap_values).shape)
```

```python
import matplotlib.pyplot as plt

# Bar plot
plt.figure(figsize=(8,5))
shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
plt.tight_layout()
plt.savefig("shap_summary_bar.png", dpi=150)
plt.show()

# Beeswarm plot
plt.figure(figsize=(10,6))
shap.summary_plot(shap_values, X_test, plot_type="dot", show=False)
plt.tight_layout()
plt.savefig("shap_beeswarm.png", dpi=150)
plt.show()
```

```python
i = 5  # test sample to explain

shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[i], X_test.iloc[i], matplotlib=True)

html = shap.force_plot(
    explainer.expected_value, shap_values[i], X_test.iloc[i]
).html()

with open(f"shap_force_index_{i}.html", "w") as f:
    f.write(html)

print(f"Saved shap_force_index_{i}.html")
```
